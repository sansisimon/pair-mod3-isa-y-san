1. ¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

```python
df = pd.read_csv('files/world_data_full_apply_nulos.csv',index_col =0)
```

```python
df.head() #birth_rate
```

```python
sns.histplot(x = "birth_rate", data = df, color = "violet", bins = 50)

# cambiamos los nombres de los ejes
plt.xlabel("tasa de natalidad")
plt.ylabel("Conteo")

# le ponemos título a la gráfica de seaborn usando el método 'plt.title()'
plt.title("Distribución de la tasa de natalidad en el conjunto de datos");
```

```python
sns.boxplot(x = "birth_rate", 
            data = df, 
            width = 0.5, 
            color = "turquoise")

# cambiamos los nombres de los ejes
plt.xlabel("tasa de natalidad")
plt.ylabel("Conteo")

# le ponemos título a la gráfica de seaborn usando el método 'plt.title()'
plt.title("Distribución de la tasa de natalidad en el conjunto de datos");
```

4. ¿Existen relaciones visibles entre la tasa de natalidad y las siguientes variables: tasa de mortalidad infantil, esperanza de vida, densidad de población y población urbana? Usa un subplot. Describe cualquier patrón que encuentres.



```python
df.head(1) # birth_rate, life_expectancy, density, infant_mortality, urban_population
```

```python
df['life_expectancy'].dtype
```

```python
for col in df[['life_expectancy', 'density', 'infant_mortality', 'urban_population']]:
    print(col)
    print(df[col].unique())
    print(df[col].dtype)
    print('-' *30)
```

```python
# birth_rate, 
# life_expectancy, density, infant_mortality, urban_population


#establecemos figura del subplot:
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (20, 5))

#primer: birth_rate, life_expectancy:
sns.regplot(x = "birth_rate", y = "life_expectancy", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "teal", "s": 1}, # cambiamos el color y el tamaño de los puntos del scaterplot
            ax = axes[0,0] # especificamos el eje en el que queremos poner la gráfica. En este caso en el eje 0. 
            )

#segundo: birth_rate, density:
sns.regplot(x = "birth_rate", y = "density", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "teal", "s": 1}, # cambiamos el color y el tamaño de los puntos del scaterplot
            ax = axes[0,1] # especificamos el eje en el que queremos poner la gráfica. En este caso en el eje 0. 
            )

#tercer: birth_rate, infant_mortality:
sns.regplot(x = "birth_rate", y = "infant_mortality", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "teal", "s": 1}, # cambiamos el color y el tamaño de los puntos del scaterplot
            ax = axes[1,0] # especificamos el eje en el que queremos poner la gráfica. En este caso en el eje 0. 
            )

#cuarto: birth_rate, urban_population:
sns.regplot(x = "birth_rate", y = "urban_population", data = df, 
            marker = "d", 
            line_kws = {"color": "black", "linewidth": 1}, # cambiamos el color y el grosor de la linea de tendencia
            scatter_kws = {"color": "teal", "s": 1}, # cambiamos el color y el tamaño de los puntos del scaterplot
            ax = axes[1,1] # especificamos el eje en el que queremos poner la gráfica. En este caso en el eje 0. 
            )
```

`conclu`: 
- esperanza de vida y natalidad tienen una relación indirecta: cuando uno sube el otro baja.
- density y urban population no tienen relación con natalidad.
- a menos mortalidad infantil, más esperanza de vida.

```python
#establecemos figura del subplot:
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (20, 5))

#primer: birth_rate, life_expectancy:
sns.scatterplot(x = "birth_rate", y = "life_expectancy", 
                data = df, 
                size = 5, 
                marker= "v",
                ax = axes[0,0])

#segundo: birth_rate, density:
sns.scatterplot(x = "birth_rate", y = "density", 
                data = df, 
                size = 5, 
                marker= "v",
                ax = axes[0,1])


#tercer: birth_rate, infant_mortality:
sns.scatterplot(x = "birth_rate", y = "infant_mortality", 
                data = df, 
                size = 5, 
                marker= "v",
                ax = axes[1,0])

#cuarto: birth_rate, urban_population:
sns.scatterplot(x = "birth_rate", y = "urban_population", 
                data = df, 
                size = 5, 
                marker= "v",
                ax = axes[1,1]);
```
